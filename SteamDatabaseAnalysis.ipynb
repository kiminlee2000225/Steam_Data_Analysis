###  Dataset 1

```python
import pandas as pd

# First dataset
df1 = pd.read_csv('steam.csv')

# Rename column name "owners" to "max_owners", "positive_ratings" to "positive_reviews", 
# "negative_ratings" to "negative_reviews"
df1.rename(columns = {"owners": "max_owners", "positive_ratings": "positive_reviews",
                    "negative_ratings": "negative_reviews"}, inplace = True)

# Return the value of max_owners to return the maximum owner, instead of the range of owners
df1["max_owners"] = df1["max_owners"].apply(lambda x: x.split("-")[1])

# Create a one-hot encoder for the supported platforms of the game
df1["support_windows"] = df1["platforms"].apply(lambda x : 1 if "windows" in x else 0)
df1["support_mac"] = df1["platforms"].apply(lambda x : 1 if "mac" in x else 0)
df1["support_linux"] = df1["platforms"].apply(lambda x : 1 if "linux" in x else 0)

# Create two new columns release_year and release_month that indicates the releade year and month of the game
release_date = pd.to_datetime(df1["release_date"])
df1["release_year"] = release_date.dt.year
df1["release_month"] = release_date.dt.month

# Create a new column "is_released_on_holiday" that determines whether the game was released on a holiday month 
# or not. The holiday months are November and December. 
df1["is_released_on_holiday"] = df1["release_month"].apply(lambda x : 1 if x == 11 or x == 12 else 0)

# Drop irrelevant columns
df1 = df1.drop(columns = {"appid", "english", "required_age", "genres", "steamspy_tags", "achievements", "platforms", 
                          "release_date", "categories", "negative_reviews", "median_playtime"})

df1.head()
```

```python
# One-way ANOVA for third hypothesis

def oneway_ANOVA(df, IV, DV):
    
    one = df[df[IV] == 1][DV]
    two = df[df[IV] == 2][DV]
    three = df[df[IV] == 3][DV]
    
    results = stats.f_oneway(one, two, three)
    f_stat = results[0]
    p_val = results[1]
    
    df1 = 3 - 1
    df2 = (len(one) - 1) + (len(two) - 1) + (len(three) - 1)
    
    print('----------------------- \n ONE-WAY ANOVA RESULTS \n----------------------- \n\nF-test \n-------')
    print('F(', df1, ",", df2, ") = ", round(f_stat, 2), ', p = ', format(p_val, '.4f'), '\n', sep="")
    
    levene_results = stats.levene(one, two, three)
    levene_true = (float(levene_results[1]) > .05)

    shapiro_one = stats.shapiro(one)
    shapiro_two = stats.shapiro(two)
    shapiro_three = stats.shapiro(three)

    if(float(shapiro_one[1]) > .05):
        one_result = 'Assumption is met. p > .05'
    else:
        one_result = 'Assumption is violated. p < .05'
    if(float(shapiro_two[1]) > .05):
        two_result = 'Assumption is met. p > .05'
    else:
        two_result = 'Assumption is violated. p < .05'
    if(float(shapiro_three[1]) > .05):
        three_result = 'Assumption is met. p > .05'
    else:
        three_result = 'Assumption is violated. p < .05'
    
    mc = MultiComparison(df[DV], df[IV])
    tukey_result = mc.tukeyhsd()
    
    print('Assumption Checks \n------------------- \nAssumption of Equality of Variances:\n\t', levene_results, sep="")
    if levene_true == True:
        print('\tAssumption is met. p > .05')
    else:
        print('\tAssumption is violated. p < .05')
    
    print('\nAssumption of Normality:\n\tOne :', shapiro_one, '\n\t\t', one_result, "\n\n\tTwo : ", 
          shapiro_two, '\n\t\t', two_result, '\n\n\tThree : ', shapiro_three, '\n\t\t', three_result)
    print('\n','Post-hoc Tests\n----------------\n', tukey_result, sep = "")
```
