```python
import pandas as pd

# First dataset
df1 = pd.read_csv('steam.csv')

# Rename column name "owners" to "max_owners", "positive_ratings" to "positive_reviews", 
# "negative_ratings" to "negative_reviews"
df1.rename(columns = {"owners": "max_owners", "positive_ratings": "positive_reviews",
                    "negative_ratings": "negative_reviews"}, inplace = True)

# Return the value of max_owners to return the maximum owner, instead of the range of owners
df1["max_owners"] = df1["max_owners"].apply(lambda x: x.split("-")[1])

#
df1["support_windows"] = df1["platforms"].apply(lambda x : 1 if "windows" in x else 0)
df1["support_mac"] = df1["platforms"].apply(lambda x : 1 if "mac" in x else 0)
df1["support_linux"] = df1["platforms"].apply(lambda x : 1 if "linux" in x else 0)

# 
release_date = pd.to_datetime(df1["release_date"])
df1["release_year"] = release_date.dt.year
df1["release_month"] = release_date.dt.month

# is_holiday season?

# Drop unnecessary columns
df1 = df1.drop(columns = {"appid", "english", "required_age", "genres", "steamspy_tags", "achievements", "platforms", 
                          "release_date", "categories"})

df1
```

```python
df1["price"].value_counts()
```

```python
# Second dataset
df2 = pd.read_csv('steam_games.csv')
df2 = df2.drop(columns = {"achievements", "game_description", "mature_content", "minimum_requirements", "recommended_requirements",
                          "discount_price", "url", "types", "desc_snippet", "game_details", "genre", "release_date",
                          "developer", "publisher", "recent_reviews", "all_reviews", "original_price"})

# Merge the first and second datasets, and set the index to the column "name"
df_final = df1.merge(df2, on = "name", how = 'inner')
df_final.set_index("name", inplace=True)


# Set target and features where the target is the positive_reviews
target_positive_reviews = df_final["positive_reviews"]
features_positive_reviews = df_final.drop("positive_reviews", axis = 1)

df_final
```

```python
popular_tags_df = df_final["popular_tags"].str.get_dummies(sep= ',')
df_final = df_final.drop(columns = {"popular_tags"})
popular_tags_df
```

```python
df_final = pd.concat([df_final, popular_tags_df], axis=1)
df_final
```

```python
languages_df = df_final["languages"].str.get_dummies(sep= ',')
df_final = df_final.drop(columns = {"languages"})
languages_df
```

```python
df_final = pd.concat([df_final, languages_df], axis=1)
df_final
```

```python
developer_df = pd.get_dummies(df_final["developer"])
df_final = df_final.drop(columns = {"developer"})
developer_df
```

```python
df_final = pd.concat([df_final, developer_df], axis=1)
df_final
```

```python
publisher_df = pd.get_dummies(df_final["publisher"])
df_final = df_final.drop(columns = {"publisher"})
publisher_df
```

```python
df_final = pd.concat([df_final, publisher_df], axis=1)
df_final
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(features_positive_reviews, target_positive_reviews, random_state=3000)


model = Lasso(alpha = 0.01).fit(X=X_train, y=y_train)


print("R-squared value for training set: ", r2_score(y_train, model.predict(X_train)))
print("R-squared value for testing set: ", r2_score(y_test, model.predict(X_test)))
```
