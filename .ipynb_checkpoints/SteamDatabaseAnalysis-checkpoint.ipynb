###  Dataset 1

```python
import pandas as pd

# First dataset
df1 = pd.read_csv('steam.csv')

# Rename column name "owners" to "max_owners", "positive_ratings" to "positive_reviews", 
# "negative_ratings" to "negative_reviews"
df1.rename(columns = {"owners": "max_owners", "positive_ratings": "positive_reviews",
                    "negative_ratings": "negative_reviews"}, inplace = True)

# Return the value of max_owners to return the maximum owner, instead of the range of owners
df1["max_owners"] = df1["max_owners"].apply(lambda x: x.split("-")[1])

# 
df1["support_windows"] = df1["platforms"].apply(lambda x : 1 if "windows" in x else 0)
df1["support_mac"] = df1["platforms"].apply(lambda x : 1 if "mac" in x else 0)
df1["support_linux"] = df1["platforms"].apply(lambda x : 1 if "linux" in x else 0)

# 
release_date = pd.to_datetime(df1["release_date"])
df1["release_year"] = release_date.dt.year
df1["release_month"] = release_date.dt.month

# Drop irrelevant columns
df1 = df1.drop(columns = {"appid", "english", "required_age", "genres", "steamspy_tags", "achievements", "platforms", 
                          "release_date", "categories"})

df1.head()
```

```python
# Shape of dataframe 1
df1.shape
```

### Dataset 2

```python
# Second dataset
df2 = pd.read_csv('steam_games.csv')

# Drop overlapping columns from the first dataframe and irrelevant columns
df2 = df2.drop(columns = {"achievements", "game_description", "mature_content", "minimum_requirements", "recommended_requirements",
                          "discount_price", "url", "types", "desc_snippet", "game_details", "genre", "release_date",
                          "developer", "publisher", "recent_reviews", "all_reviews", "original_price"})

df2.head()
```

```python
# Shape of dataframe 2
df2.shape
```

## Merging the Datasets

```python
# Merge the first and second dataframes, and set the index to the column "name"
df_final = df1.merge(df2, on = "name", how = 'inner')
df_final.set_index("name", inplace=True)

df_final.head()
```

```python
# Shape of the merged dataframe
#
# Number of columns should be 15 since (14 + 3 - 1 - 1)
# where 14 is the number of columns for df1, 3 is the number of columns for df2, the first - 1 is 
# caused by the merging of the two dataframes on the column "name", and the second -1 is caused by
# column "name" becoming an index. 
df_final.shape
```

## One-Hot Encoding

We will be one hot encoding the columns popular_tags, languages, developer, and publisher. Each of these columns include categorical data, which needs to be altered into values of 0 or 1. We need to change them into a numeric variable so that the common machine learning algorithms can be applied effectively. Each of the columns provide an attribute of the game that can be used as a feature to predict the target variable. 

### Popular Tags

```python
# Create a dataframe for the popular tags by splitting the tags in the column "popular_tags" 
# into dummy variables. Then delete the column in the merged dataframe. Use the same method for 
# the languages column. 
popular_tags_df = df_final["popular_tags"].str.get_dummies(sep= ',')
df_final = df_final.drop(columns = {"popular_tags"})
popular_tags_df.head()
```

```python
# Shape of the popular_tags_df
# Row represents the number of unique games
# Column represents the number of unique tags used by Steam for any game
popular_tags_df.shape
```

```python
# Merge popular_tags_df to the final dataset
df_final = pd.concat([df_final, popular_tags_df], axis=1)
df_final.head()
```

### Languages

```python
languages_df = df_final["languages"].str.get_dummies(sep= ',')
df_final = df_final.drop(columns = {"languages"})
languages_df.head()
```

```python
# Shape of the languages_df
# Row represents the number of unique games
# Column represents the number of unique languages supported for any game
languages_df.shape
```

```python
# Merge languages_df to the final dataset
df_final = pd.concat([df_final, languages_df], axis=1)
df_final.head()
```

### Developer

```python
# Create a dataframe for the developers by splitting the tags in the column "developer" 
# into dummy variables. Then delete the column in the merged dataframe. Use the same method for 
# the publisher column. 
developer_df = pd.get_dummies(df_final["developer"])
df_final = df_final.drop(columns = {"developer"})
developer_df.head()
```

```python
# Shape of the developer_df
# Row represents the number of unique games
# Column represents the number of unique developers for any game
developer_df.shape
```

```python
# Merge developer_df to the final dataset
df_final = pd.concat([df_final, developer_df], axis=1)
df_final.head()
```

### Publisher

```python
publisher_df = pd.get_dummies(df_final["publisher"])
df_final = df_final.drop(columns = {"publisher"})
publisher_df.head()
```

```python
# Shape of the publisher_df
# Row represents the number of unique games
# Column represents the number of unique publishers for any game
publisher_df.shape
```

```python
# Merge publisher_df to the final dataset
df_final = pd.concat([df_final, publisher_df], axis=1)
df_final.head()
```

```python
# Set target and features where the target is the positive_reviews
target_positive_reviews = df_final["positive_reviews"]
features_positive_reviews = df_final.drop("positive_reviews", axis = 1)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(features_positive_reviews, target_positive_reviews, random_state=3000)


model = Lasso(alpha = 0.01).fit(X=X_train, y=y_train)


print("R-squared value for training set: ", r2_score(y_train, model.predict(X_train)))
print("R-squared value for testing set: ", r2_score(y_test, model.predict(X_test)))
```
